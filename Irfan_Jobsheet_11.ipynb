# Praktikum 1

```python
def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return f"Data {target} ditemukan di indeks {i}"
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

# Praktikum 2

```python
def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Data {target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

# Praktikum 5

```python
import time

def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return i
  return -1

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return mid
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return -1

data = list(range(1, 100001))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print("Sequential Search waktu:", end - start, "detik")

start = time.time()
binary_search(data, target)
end = time.time()
print("Binary Search waktu:", end - start, "detik")
```

# Praktikum 6

```python
import random

def insertion_sort(data):
  for i in range(1, len(data)):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"{target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"{target} tidak ditemukan"

data = [random.randint(1, 100) for _ in range(15)]
print("Data acak sebelum diurutkan:", data)
urut = insertion_sort(data)
print("Data setelah diurutkan:", urut)
target = urut[7]
print(binary_search(urut, target))
```

# Penugasan 2

```python
def binary_search_v2(data, target):
    left, right = 0, len(data) - 1
    iterasi = 0

    while left <= right:
        iterasi += 1
        mid = (left + right) // 2
        if data[mid] == target:
            return f"Data {target} ditemukan di indeks {mid} (Jumlah iterasi: {iterasi})"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return f"Data {target} tidak ditemukan (Jumlah iterasi: {iterasi})"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search_v2(angka, 40))
print(binary_search_v2(angka, 100))
```

# Penugasan 5

```python
def analisis_kompleksitas():
    ukuran_data = [100, 1000, 10000]
    print(f"\n{'Ukuran':<10} | {'Seq Search (Waktu)':<20} | {'Bin Search (Waktu)':<20}")
    print("-" * 60)

    for n in ukuran_data:
        data = sorted(random.sample(range(n * 10), n))
        target = data[-1]

        start = time.perf_counter()
        sequential_search(data, target)
        end = time.perf_counter()
        t_seq = end - start

        start = time.perf_counter()
        binary_search(data, target)
        end = time.perf_counter()
        t_bin = end - start

        print(f"{n:<10} | {t_seq:<20.6f} | {t_bin:<20.6f}")

analisis_kompleksitas()
```

# Penugasan 6

```python
def proyek_mini_nilai():
    print("--- Proyek Mini: Pencarian Nilai Mahasiswa ---")

    data_mahasiswa = []
    jumlah = int(input("Masukkan jumlah mahasiswa yang ingin diinput: "))

    for i in range(jumlah):
        nama = input(f"Masukkan nama mahasiswa ke-{i+1}: ")
        nilai = int(input(f"Masukkan nilai untuk {nama}: "))
        data_mahasiswa.append((nama, nilai))

    data_mahasiswa.sort(key=lambda x: x[1])

    print("\nData Mahasiswa Terurut (berdasarkan nilai):")
    for mhs in data_mahasiswa:
        print(f"- {mhs[0]}: {mhs[1]}")

    target_nilai = int(input("\nMasukkan nilai yang ingin dicari: "))

    left, right = 0, len(data_mahasiswa) - 1
    iterasi = 0
    ditemukan = False

    while left <= right:
        iterasi += 1
        mid = (left + right) // 2
        if data_mahasiswa[mid][1] == target_nilai:
            print(f"Data nilai {target_nilai} ditemukan pada mahasiswa: {data_mahasiswa[mid][0]} (Indeks: {mid}, Iterasi: {iterasi})")
            ditemukan = True
            break
        elif data_mahasiswa[mid][1] < target_nilai:
            left = mid + 1
        else:
            right = mid - 1

    if not ditemukan:
        print(f"Data nilai {target_nilai} tidak ditemukan (Jumlah iterasi: {iterasi})")

proyek_mini_nilai()
```
