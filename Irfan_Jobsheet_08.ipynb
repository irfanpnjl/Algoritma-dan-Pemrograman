# Praktikum 04 - Fungsi dengan *args dan **kwargs

```python
def jumlahkan (*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print("total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

# Praktikum 07 - Fungsi decorator

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan...")
    fungsi()
    print("Setelah fungsi dijalankan...")
  return wrapper

@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines!")
sapa()
```

# Praktikum 08 - Studi Kasus: Menghitung Rata-rata Nilai Mahasiswa

```python
def hitung_rata(data):
  return sum(data) / len(data)

nilai = [80, 90, 85]
rata = hitung_rata(nilai)

print("Rata-rata nilai: ", rata)
```

# Penugasan - args

# Penugasan - kwargs

```python
def info_barang(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

info_barang(Nama="Buku", Harga=15000, Stok=30)
```

# Penugasan - Lambda

```python
kuadrat = lambda x, y: x ** y
print("Hasil kuadrat:", kuadrat(4, 2))

kel_lingkaran = lambda r: 2 * 3.14 * r
print("Keliling lingkaran:", kel_lingkaran(7))
```

# Penugasan - Decorator

```python
def logger (fungsi):
  def wrapper():
    print("Memulai fungsi. . .")
    fungsi()
    print("Fungsi selesai dijalankan.")
  return wrapper

@logger
def halo():
  print("Halo, Nama saya Irfan")

halo()
```
