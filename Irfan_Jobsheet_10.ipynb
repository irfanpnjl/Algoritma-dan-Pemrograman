# Praktikum 01

```python
def bubble_sort(data):
  n = len(data)
  for i in range (n):
    for j in range(0, n - i - 1):
      if data[j] > data[j + 1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

# Praktikum 02

```python
def selection_sort(data):
  n = len(data)
  for i in range(n):
    min_index = 1
    for j in range(i + 1, n):
      if data[j] < data[min_index]:
        min_index = j
    data[i], data[min_index] = data[min_index], data[i]
  return data

angka = [64, 25, 12, 22,11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

# Praktikum 03

```python
def insertion_sort(data):
  n = len(data)
  for i in range(1, n):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data [j]
      j -= 1
    data[j + 1] = key
  return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

# Praktikum 06

```python
def bubble_sort_key(data, key_index=0, reverse=False):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if reverse:
        if data[j][key_index] > data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
        else:
          if data[j][key_index] > data[j + 1][key_index]:
            data[j], data[j + 1] = data[j + 1], data[j]

  return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000),("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(),0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(),1, reverse=True))
```

# Penugasan 5

```python
import random
def count_comparisons(func, data):
    comparisons = 0
    arr = data.copy()
    n = len(arr)

    if func == "bubble":
        for i in range(n):
            for j in range(0, n-i-1):
                comparisons += 1
                if arr[j] > arr[j+1]:
                    arr[j], arr[j+1] = arr[j+1], arr[j]
    elif func == "selection":
        for i in range(n):
            min_idx = i
            for j in range(i+1, n):
                comparisons += 1
                if arr[j] < arr[min_idx]:
                    min_idx = j
            arr[i], arr[min_idx] = arr[min_idx], arr[i]

    return comparisons

sizes = [10, 50, 100]
for size in sizes:
    test_data = [random.randint(1, 1000) for _ in range(size)]
    print(f"Ukuran Data: {size}")
    print(f"- Bubble: {count_comparisons('bubble', test_data)} perbandingan")
    print(f"- Selection: {count_comparisons('selection', test_data)} perbandingan")
```

# Penugasan 6

```python
def sort_barang(daftar, kunci, descending=False):
    n = len(daftar)
    for i in range(n):
        for j in range(0, n-i-1):
            kondisi = daftar[j][kunci] < daftar[j+1][kunci] if descending else daftar[j][kunci] > daftar[j+1][kunci]
            if kondisi:
                daftar[j], daftar[j+1] = daftar[j+1], daftar[j]

daftar_barang = [
    {"nama": "Laptop", "harga": 15000000},
    {"nama": "Mouse", "harga": 200000},
    {"nama": "Keyboard", "harga": 500000},
    {"nama": "Monitor", "harga": 2500000}
]

print("Urutkan berdasarkan Harga (Descending):")
sort_barang(daftar_barang, "harga", descending=True)
for b in daftar_barang:
    print(f"{b['nama']}: Rp{b['harga']}")
```
